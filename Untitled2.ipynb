### ЗАДАЧИ ЛИНЕЙНОГО ПРОГРАММИРОВАНИЯ (ДВОЙКА - 3, ВАРИАНТ - 3)

Найти такие числа $x$ и $y$, удовлетворяющие условиям:

\
\begin{cases}
y \geq 0 \\
x + y \leq 1 \\
x - 4y \geq -2
\end{cases}


при которых разность $y - x$ принимает наибольшее значение.

###ИМПОРТ БИБЛИОТЕК И ПОСТАНОВКА ЗАДАЧИ

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
import pulp
from cvxopt import matrix, solvers
import time

# Постановка задачи
print("ЗАДАЧА 3 (Вариант 3)")
print("Максимизировать: y - x")
print("Ограничения:")
print("1. y ≥ 0")
print("2. x + y ≤ 1")
print("3. x - 4y ≥ -2")
print("\nОжидаемый ответ: x = -2; y = 0")
```

###РЕШЕНИЕ С ПОМОЩЬЮ SCIPY.OPTIMIZE

```python
start_scipy = time.time()

# Преобразуем задачу максимизации в минимизацию
# Максимизировать y - x = минимизировать x - y
c = [1, -1]  # коэффициенты для x - y

# Ограничения в форме A_ub * x <= b_ub
# y ≥ 0 → -y ≤ 0
# x + y ≤ 1
# x - 4y ≥ -2 → -x + 4y ≤ 2
A_ub = [
    [0, -1],   # -y ≤ 0
    [1, 1],    # x + y ≤ 1
    [-1, 4]    # -x + 4y ≤ 2
]
b_ub = [0, 1, 2]

# Границы переменных
x_bounds = (None, None)  # x не ограничен
y_bounds = (0, None)     # y ≥ 0

# Решаем задачу
result_scipy = linprog(c, A_ub=A_ub, b_ub=b_ub,
                      bounds=[x_bounds, y_bounds],
                      method='simplex')

time_scipy = time.time() - start_scipy

print("РЕЗУЛЬТАТЫ SCIPY:")
if result_scipy.success:
    x_opt_scipy = result_scipy.x[0]
    y_opt_scipy = result_scipy.x[1]
    max_value_scipy = y_opt_scipy - x_opt_scipy
    print(f"✓ x = {x_opt_scipy:.6f}")
    print(f"✓ y = {y_opt_scipy:.6f}")
    print(f"✓ Максимальное значение y - x = {max_value_scipy:.6f}")
else:
    print("✗ Решение не найдено")

print(f"⏱️ Время решения: {time_scipy:.6f} секунд")
```

###РЕШЕНИЕ С ПОМОЩЬЮ PULP

```python
start_pulp = time.time()

# Создаем задачу максимизации
problem_pulp = pulp.LpProblem("Maximize_y_minus_x", pulp.LpMaximize)

# Определяем переменные
x = pulp.LpVariable("x", lowBound=None)  # x не ограничен снизу
y = pulp.LpVariable("y", lowBound=0)     # y ≥ 0

# Целевая функция: максимизировать y - x
problem_pulp += y - x, "Objective_Function"

# Ограничения
problem_pulp += x + y <= 1, "Constraint_1"
problem_pulp += x - 4*y >= -2, "Constraint_2"

# Решаем задачу
problem_pulp.solve()

time_pulp = time.time() - start_pulp

print("РЕЗУЛЬТАТЫ PULP:")
if problem_pulp.status == pulp.LpStatusOptimal:
    x_opt_pulp = pulp.value(x)
    y_opt_pulp = pulp.value(y)
    max_value_pulp = pulp.value(y - x)
    print(f"✓ x = {x_opt_pulp:.6f}")
    print(f"✓ y = {y_opt_pulp:.6f}")
    print(f"✓ Максимальное значение y - x = {max_value_pulp:.6f}")
else:
    print("✗ Решение не найдено")

print(f"⏱️ Время решения: {time_pulp:.6f} секунд")
```

###РЕШЕНИЕ С ПОМОЩЬЮ CVXOPT

```python
start_cvxopt = time.time()

# CVXOPT решает задачи минимизации, поэтому преобразуем:
# Максимизировать y - x = минимизировать x - y
c_cvxopt = matrix([1.0, -1.0])  # коэффициенты для x - y

# Ограничения в форме G*x <= h
# -y ≤ 0 → [0, -1] * [x; y] ≤ 0
# x + y ≤ 1 → [1, 1] * [x; y] ≤ 1
# x - 4y ≥ -2 → -x + 4y ≤ 2 → [-1, 4] * [x; y] ≤ 2
G = matrix([[0.0, 1.0, -1.0],
            [-1.0, 1.0, 4.0]])
h = matrix([0.0, 1.0, 2.0])

# Решаем задачу (отключаем вывод)
solvers.options['show_progress'] = False
sol = solvers.lp(c_cvxopt, G, h)

time_cvxopt = time.time() - start_cvxopt

print("РЕЗУЛЬТАТЫ CVXOPT:")
if sol['status'] == 'optimal':
    x_opt_cvxopt = sol['x'][0]
    y_opt_cvxopt = sol['x'][1]
    max_value_cvxopt = y_opt_cvxopt - x_opt_cvxopt
    print(f"✓ x = {x_opt_cvxopt:.6f}")
    print(f"✓ y = {y_opt_cvxopt:.6f}")
    print(f"✓ Максимальное значение y - x = {max_value_cvxopt:.6f}")
else:
    print("✗ Решение не найдено")

print(f"⏱️ Время решения: {time_cvxopt:.6f} секунд")
```

###СВОДНАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ

```python
print(f"{'Метод':<15} {'x':<10} {'y':<10} {'y-x':<12} {'Время (с)':<10}")
print("-" * 60)

methods = [
    ("Scipy", x_opt_scipy, y_opt_scipy, max_value_scipy, time_scipy),
    ("Pulp", x_opt_pulp, y_opt_pulp, max_value_pulp, time_pulp),
    ("CVXOPT", x_opt_cvxopt, y_opt_cvxopt, max_value_cvxopt, time_cvxopt)
]

for method, x_val, y_val, value, time_val in methods:
    if x_val is not None:
        print(f"{method:<15} {x_val:<10.6f} {y_val:<10.6f} {value:<12.6f} {time_val:<10.6f}")
    else:
        print(f"{method:<15} {'-':<10} {'-':<10} {'-':<12} {time_val:<10.6f}")
```

###ГРАФИЧЕСКОЕ РЕШЕНИЕ

```python
# Создаем график
fig, ax = plt.subplots(figsize=(12, 10))

# Определяем область построения
x_vals = np.linspace(-3, 2, 400)

# Строим ограничения
# 1. y >= 0
ax.axhline(y=0, color='red', linestyle='-', linewidth=2, label='y ≥ 0')

# 2. x + y <= 1 → y <= 1 - x
y1 = 1 - x_vals
ax.plot(x_vals, y1, 'green', linestyle='-', linewidth=2, label='x + y ≤ 1')

# 3. x - 4y >= -2 → y <= (x + 2)/4
y2 = (x_vals + 2) / 4
ax.plot(x_vals, y2, 'blue', linestyle='-', linewidth=2, label='x - 4y ≥ -2')

# Заполняем допустимую область
y_upper = np.minimum(1 - x_vals, (x_vals + 2) / 4)
y_upper = np.maximum(y_upper, 0)  # учитываем y >= 0

ax.fill_between(x_vals, 0, y_upper, where=(y_upper >= 0),
                alpha=0.3, color='lightblue', label='Допустимая область')

print("Построены ограничения и допустимая область")
```

###АНАЛИЗ ВЕРШИН И ОПТИМАЛЬНОГО РЕШЕНИЯ

```python
# Находим вершины допустимой области
vertices = []

# Вершина A: пересечение y = 0 и x - 4y = -2
x_a = -2
y_a = 0
vertices.append((x_a, y_a, 'A'))

# Вершина B: пересечение y = 0 и x + y = 1
x_b = 1
y_b = 0
vertices.append((x_b, y_b, 'B'))

# Вершина C: пересечение x + y = 1 и x - 4y = -2
# Решаем систему: x + y = 1, x - 4y = -2
x_c = 0.4
y_c = 0.6
vertices.append((x_c, y_c, 'C'))

# Вычисляем целевую функцию в вершинах
print("АНАЛИЗ ВЕРШИН ДОПУСТИМОЙ ОБЛАСТИ:")
max_vertex_value = -float('inf')
optimal_vertex = None

for x_v, y_v, name in vertices:
    value = y_v - x_v
    print(f"Вершина {name}: ({x_v}, {y_v}), y - x = {value}")

    # Отмечаем вершину на графике
    ax.plot(x_v, y_v, 'ko', markersize=8, label=f'Вершина {name}({x_v}, {y_v})')

    if value > max_vertex_value:
        max_vertex_value = value
        optimal_vertex = (x_v, y_v, name)

print(f"\n ОПТИМАЛЬНАЯ ВЕРШИНА: {optimal_vertex[2]}{optimal_vertex[0:2]}")
print(f" МАКСИМАЛЬНОЕ ЗНАЧЕНИЕ y - x = {max_vertex_value}")

# Отмечаем оптимальную точку
ax.plot(optimal_vertex[0], optimal_vertex[1], 'ro', markersize=12,
        label=f'ОПТИМУМ {optimal_vertex[2]}({optimal_vertex[0]}, {optimal_vertex[1]})')
```

###ЛИНИИ УРОВНЯ И ГРАДИЕНТ

```python
# Строим линии уровня целевой функции y - x = const
levels = [-1, 0, 1, 2]
colors = ['lightgray', 'gray', 'darkgray', 'black']

for level, color in zip(levels, colors):
    y_level = x_vals + level
    ax.plot(x_vals, y_level, color=color, linestyle='--', alpha=0.7,
            label=f'y - x = {level}' if level == 2 else "")

# Строим градиент (направление увеличения целевой функции)
# Для y - x градиент: (-1, 1)
ax.arrow(-1, 0.5, -0.5, 0.5, head_width=0.05, head_length=0.1,
         fc='purple', ec='purple', linewidth=3, label='Градиент (-1, 1)')

print("Построены линии уровня и градиент")
```

###ФИНАЛЬНОЕ ОФОРМЛЕНИЕ ГРАФИКА

```python
# Настройки графика
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('Графическое решение задачи линейного программирования\n'
             'Максимизация: y - x\n'
             f'Оптимальное решение: x = {optimal_vertex[0]}, y = {optimal_vertex[1]}, y - x = {max_vertex_value}',
             fontsize=14, pad=20)

ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_xlim([-3, 2])
ax.set_ylim([-0.5, 1.5])
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Добавляем координатную сетку
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
